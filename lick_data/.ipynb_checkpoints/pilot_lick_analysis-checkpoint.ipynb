Analysis of licking data for two C57BL6 female mice injected with AAV-DJ-CaMKII-GCaMP6s and AAV-DJ-CaMKII-mCherry into the left dorsal hippocampus.

Task is a 2 alternative forced choice paradigm, where the VR cues indicate which of two lickports will be rewarded.
Rewards are dispensed regardless of the mouse's behavior at this point. 
I will soon be adding the requirement that the mouse licks in order to receive the reward. I wanted to see if anticipitory licking arose spontaneously

```python


def align_rewards_licks(rf,lf,stepSize=.05):
    interpLicks = interpolate_licks(lf)
    postLicksT = 1.5
    nsteps = int(np.ceil(postLicksT/stepSize))
    lickDat = np.zeros([rf.shape[0],interpLicks.shape[1],nsteps])
    for row in range(rf.shape[0]):
        rewardStartInd = np.abs(interpLicks[:,-1]-rf[row,1])
        firstInd = np.argmin(rewardStartInd)
        rowDat = interpLicks[firstInd:firstInd+nsteps,:].transpose()
        lickDat[row,:,:rowDat.shape[1]] = interpLicks[firstInd:firstInd+nsteps,:].transpose()
        
    return lickDat

def interpolate_licks(lf):
    '''interpolate frame times of licks to get an even grid'''
    stepSize = .05
    rawTime = lf[:,3]
    rawLicksL, rawLicksR = lf[:,0], lf[:,1]
    rawPos = lf[:,2]
    time = np.arange(np.floor(rawTime[0]),np.ceil(rawTime[-1]),stepSize)
    
    interpDat = np.zeros([time.size-1, 4])
    for i in range(time.size-1):
        inds = np.where((rawTime>= time[i]) & (rawTime<=time[i+1]))
        
        if inds[0].size == 0:
            interpDat[i,:] = np.array([0,0,np.nan, time[i]])
        else:
            interpDat[i,:] = np.array([rawLicksL[inds[0]].sum(), rawLicksR[inds[0]].sum(), 
                                        rawPos[inds[0]].mean(), time[i]])
    return interpDat
    

```

```python
#mice = ['3984648.1', '3984648.2']
mice = ['3984673.0']
basename = '/Volumes/mplitt/VR/'

# most recent sessions only
#sessions = ['train_7','train_8','train_9','train_10','train_11','scan0']
sessions = ['1_LR']
#sessions = ['prescan1', 'scan1']

data = {}
for i in range(len(mice)):
    tmp_dat = {}
    mouse = mice[i]
    #sessions = glob.glob(basename+mouse+'/*_licks.txt')
    lickList, rewardList = [ ], [ ]
    LLicks = 0.*np.arange(0,1.5,.05)
    RLicks = 0.*np.arange(0,1.5,.05)
    for sess in sessions:
        stem = basename+mouse+'/'+sess
        lickFile = np.loadtxt(stem+'_licks.txt',delimiter = '\t')
        #L licks    R Licks   position    time   morph
        rewardFile = np.loadtxt(stem + '_rewards.txt',delimiter = '\t',converters={3: lambda s: int(s in ["True"]) })
        # position   time_since_start    morph    side of reward (True == left, False == right)
        
        # call align_rewards_licks
        lickTensor = align_rewards_licks(rewardFile,lickFile)
        lickList.append(lickTensor)
        rewardList.append(rewardFile)
        
        # reward ind x (L licks    R Licks   position    time ) x num steps
        
    tmp_dat['lickList'] = lickList
    tmp_dat['rewardList']=rewardList
    tmp_dat['sessions'] = sessions
    data[mouse] = tmp_dat
    
    
        
    
```

```python
fig,axarr = plt.subplots(2, 1)
fig2, ax2arr = plt.subplots(2,1)
fig3, ax3arr = plt.subplots(2,1)

zeroVec = 0.*np.arange(0,1.5,.05)
i = 0
t=np.arange(0,1.5,.05)
for mouse in data.keys():
    LLicks, RLicks = 0.*np.arange(0,1.5,.05), 0.*np.arange(0,1.5,.05)
    LLick_LTrial, RLick_LTrial = 0.*np.arange(0,1.5,.05), 0.*np.arange(0,1.5,.05)
    LLick_RTrial, RLick_RTrial = 0.*np.arange(0,1.5,.05), 0.*np.arange(0,1.5,.05)
    dat = data[mouse]
    for sess in range(len(dat['sessions'])):
        #print(dat['rewardList'][sess][:,3])
        LTrialInds = np.where(dat['rewardList'][sess][:,2] == 1.)
        RTrialInds = np.where(dat['rewardList'][sess][:,2] == 0)
        
        LLicks += np.mean(dat['lickList'][sess][:,0,:],axis=0)
        RLicks += np.mean(dat['lickList'][sess][:,1,:],axis=0)
        
        LLick_LTrial += dat['lickList'][sess][LTrialInds[0],0,:].mean(axis=0)
        RLick_LTrial += dat['lickList'][sess][LTrialInds[0],1,:].mean(axis=0)
        
        LLick_RTrial += dat['lickList'][sess][RTrialInds[0],0,:].mean(axis=0)
        RLick_RTrial += dat['lickList'][sess][RTrialInds[0],1,:].mean(axis=0)
        
        
        if sess > 0:
            LLicks /= 2.
            RLicks /= 2.
            
            LLick_LTrial /=2.
            RLick_LTrial /= 2.
            
            LLick_RTrial /= 2.
            RLick_RTrial /= 2.
            
    
    axarr[i].plot(t,LLicks)
    axarr[i].plot(t,RLicks)
    
    axarr[i].vlines(.5,0,1)
    
    ax2arr[i].plot(t,LLick_LTrial)
    ax2arr[i].plot(t,RLick_LTrial)
    ax2arr[i].vlines(.5,0,1)
    
    ax3arr[i].plot(t,LLick_RTrial)
    ax3arr[i].plot(t,RLick_RTrial)
    ax3arr[i].vlines(.5,0,1)
    i+=1
    
axarr[0].set_title("Overall Licking behavior\n mouse 1")
axarr[1].set_title("mouse 2")
axarr[1].legend(('left','right'))
axarr[0].set_ylabel("Lick Rate")
axarr[1].set_xlabel("time (s)")
fig.subplots_adjust(hspace=.5)
    
ax2arr[0].set_title("Left Reward Trials \n mouse 1")
ax2arr[1].set_title("mouse 2")
fig2.subplots_adjust(hspace=.5)

ax3arr[0].set_title("Right Reward Trials \n mouse 1")
ax3arr[1].set_title("mouse 2")
fig3.subplots_adjust(hspace=.5)
        
```

Time 0 corresponds to the onset of the reward cue. The vertical black line above indicates onset of reward.

Clearly mouse 1's tactic is just to lick right and then lick left if the reward doesn't show up. I think this confirms that I need to require the mice actively lick to the side that they think the reward should be. 

Mice seem to develop spontaneous anticipitory licking though! This is perhaps due to the explicit reward cue.
