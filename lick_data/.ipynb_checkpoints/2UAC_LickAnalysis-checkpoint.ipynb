```python
import numpy as np
from matplotlib import pyplot as plt
import scipy as sp
from behavioral_data import process_data as pd
import behavioral_data as bd

%matplotlib inline

mice= ['3984648.2','3984672.1','3984672.2']
#mice = ['3984672.2']
```

```python
for mouse in mice:
    print(mouse)
    c = pd(mouse,[])
    #c.save_sessions()
    R,D = c.load_sessions()
    bd.plot_learning_curve(R,title=mouse)
    
```

```python
for mouse in mice:
    print(mouse)
    c = pd(mouse,[])
    #c.save_sessions()
    R,D = c.load_sessions()
    sessList = sorted([int(i.split('_')[0]) for i in c.sessions])
    sessions = [str(i)+'_LR' for i in sessList]
    for sess in sessions:
        #print(sess)
        bd.reward_trig_lick_rate(R[sess],title=mouse+'_'+sess)
        mv, pRight,RT, omit = bd.psychometrics(R[sess],title=mouse+'_'+sess,plots='all')
    
```
